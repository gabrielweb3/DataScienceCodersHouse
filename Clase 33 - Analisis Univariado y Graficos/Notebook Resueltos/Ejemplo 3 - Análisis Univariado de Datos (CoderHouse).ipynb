```python
import pandas as pd
reviews = pd.read_csv("C:/Users/Layla/Desktop/Docencia/Coder House/Materiales complementarios/Mod 4 - Data Science/3. Análisis Univariado/Datos/winemag-data_first150k.csv", index_col=0)
reviews.head()
```

```python
# Gráficos de barras y datos categóricos
reviews['province'].value_counts().head(10).plot.bar()
```

¿Qué nos dice este grafico? Se observa que California produce mucho más vino que cualquier otra provincia del mundo! Podríamos preguntarnos qué porcentaje del total es cosecha californiana. Este gráfico de barras nos dice números absolutos, pero es más útil conocer las proporciones relativas.

```python
(reviews['province'].value_counts().head(10) / len(reviews)).plot.bar()
```

El grafico anterior demuestra que California produce casi un tercio de los vinos reseñados en Wine Magazine!
