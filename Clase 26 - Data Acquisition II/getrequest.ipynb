```python
import requests
import json
import pandas as pd

url = 'https://jsonplaceholder.typicode.com/users'

respuesta = requests.get(url)
type(respuesta)
```

```python
type(respuesta.json())
```

```python
respuesta_dict = json.loads(respuesta.text)
respuesta_dict
```

```python
pd.DataFrame.from_dict(respuesta_dict)
```
